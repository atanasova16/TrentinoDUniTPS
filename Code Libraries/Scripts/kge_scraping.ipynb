```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
html = list(soup.children)[2]
```

```python
results = soup.find(id="du-content")
```

```python
results.prettify()
```

```python
departments = results.find_all("div", class_="col-sm-4")
```

```python
for dep in departments:
  print(dep, end="\n"*2)
```

```python
departments_info = []

for dep in departments:
    dep_info = {}
    title_element = dep.find("div", class_="title")
    director_element = dep.find("div", class_="direttore")
    address_element = dep.find("div", class_="indirizzo")
    website_element = dep.find("div", class_="sito-web")
    #print(title_element)
    name = title_element.find('h2').text #this is the department's name
    #print(director_element)
    director = director_element.text[11:]
    #print(address_element)
    address = address_element.text
    #print(website_element)
    website = website_element.text
    dep_info["name_dep"] = name
    dep_info["director"] = director
    dep_info["address"] = address
    dep_info["website"] = website
    print(dep_info)
    departments_info.append(dep_info)
    print()
```
