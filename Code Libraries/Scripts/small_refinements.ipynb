```python
#df.loc[df[<some_column_name>] == <condition>, [<another_column_name>]] = <value_to_add>

#departments_data.insert(0,"department_id", " ")
for dep_name in departments_data['name_dep']:
    #print(dep_name)
    # if dep_name in deg_dep['department_name'].values:
    #     print("YES")
    #     unitid = deg_dep[deg_dep['department_name' == dep_name]]
    if dep_name in deg_dep['department_name'].values:
        #print(dep_name)
        #print(deg_dep.loc[deg_dep['department_name'] == dep_name].iloc[0].department_id)
        dep_id = deg_dep.loc[deg_dep['department_name'] == dep_name].iloc[0].department_id
        departments_data.loc[departments_data['name_dep'] == dep_name, 'department_id'] = dep_id
        
    else:
        print("=================")
        print(dep_name)
```

## Add id to degrees

```python
degrees_data = pd.read_csv("C:/Users/User/Documents/UNITN/KGE/project_datasets/degrees_departments.csv")
degrees_data
```

```python
degrees_data['degree_id'] = "id"
degrees_data
```

```python
degrees_data = degrees_data.rename(columns = {'corsoStudi': 'degree_name'})
degrees_data
```

```python
j = 1
for i in range(len(degrees_data)):
    value = degrees_data.loc[i, "degree_name"].replace(" ", '').lower() + '_' + str(j)
    degrees_data.loc[i, "degree_id"] = value
    j += 1
degrees_data
```

```python
degrees_data.to_csv("C:/Users/User/Documents/UNITN/KGE/project_datasets/degrees_departments.csv", index = False)
```

## Extracting degree type

```python
degrees_data = pd.read_csv("C:/Users/User/Documents/UNITN/KGE/project_datasets/degrees_departments.csv")
degrees_data
```

```python
degrees_data['degree_type'] = "laurea"
```

```python

for i in range(len(degrees_data)):
    name = degrees_data.loc[i, "degree_name"]
    idx = name.rfind("(")
    value = name[idx+1:-1]
    #print(value)
    degrees_data.loc[i, "degree_type"] = value
    
degrees_data
```
